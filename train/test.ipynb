```python
import keras
from random import randint
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMG_SIZE = (224, 224)
dataDir = "datasets/processed_binary"
```

```python
seed = randint(0, 2**32 - 1)
```

```python
base_model = keras.applications.MobileNetV2(
    input_shape=IMG_SIZE + (3,), include_top=True, weights="imagenet"
)
```

```python
model = base_model
```

```python
img = Image.open(
    "datasets/processed_binary/close/Camera2_23-09-53_mp4-18_jpg.rf.75c29333c637bb984a38179fc78f0b73.jpg"
)
```

```python
plt.imshow(img)
```

```python
img = img.resize(IMG_SIZE)
```

```python
batch = np.empty((1,)+IMG_SIZE+(3,), dtype=np.float32)
batch[0] = img
```

```python
model_input = keras.applications.mobilenet_v2.preprocess_input(batch)
```

```python
pred = model.predict(model_input)
```

```python
keras.applications.mobilenet_v2.decode_predictions(pred)
```
