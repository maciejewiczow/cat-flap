```python
from PIL import Image, ImageDraw
from matplotlib import pyplot as plt
import numpy as np
from ultralytics import YOLO
```

```python
imagePath = "datasets/processed_binary/close/Camera2_23-09-53_mp4-18_jpg.rf.75c29333c637bb984a38179fc78f0b73.jpg"
```

```python
img = Image.open(imagePath)
```

```python
plt.imshow(img)
```

```python
model = YOLO("yolo12n.pt")
```

```python
results = model(imagePath)
```

```python
annotated = ImageDraw.Draw(img.copy())

for result in results:
    xyxy = result.boxes.xyxy

    annotated.rectangle([(xyxy[0,0], xyxy[0,1]), (xyxy[0,2], xyxy[0,3])], fill=None, outline='red', width=8)

    # [result.names[cls.item()] for cls in result.boxes.cls.int()]

plt.imshow(annotated._image)
```
